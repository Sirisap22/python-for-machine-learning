# Relational Plot

## sns.scatterplot

```python
sns.scatterplot?
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')

x = np.linspace(0, 50, 100)
y = x * 0.10
plt.plot(x, y)
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', palette='winter')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='size', size='size')
```

## sns.lineplot

```python
x = np.linspace(0, 10, 100)
y = x ** 2
sns.lineplot(x=x, y=y)
```

## sns.relplot

```python
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', col='sex', hue='smoker')
```

```python
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', row='sex', col='day', hue='smoker')
```

```python
sns.relplot(
    kind='scatter', data=tips, 
    x='total_bill', y='tip', row='sex', 
    col='day', hue='smoker', 
    height=4, aspect=1.2)
```

## sns.jointplot

```python
sns.jointplot(kind='scatter', data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(kind='hex', data=tips, x='total_bill', y='tip')
```

## sns.lmplot

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', col='time')
```

# Distribution Plots

## sns.histplot

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.histplot(data=tips['total_bill'], bins=20)
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, hue='sex')
```

## sns.kdeplot

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='day')
```

## sns.ecdfplot

```python
sns.ecdfplot(data=tips, x='tip')
```

```python
sns.ecdfplot(data=tips, x='tip', hue='smoker')
```

## sns.displot

```python
sns.displot(kind='hist', data=tips, x='tip', col='day', row='sex', hue='sex')
```

```python
sns.displot(kind='kde', data=tips, x='tip', col='day', row='sex', hue='sex')
```

# Pair Plot

```python
tips.info()
```

```python
sns.pairplot(data=tips)
```

# Categorical Plots

## sns.barplot
[Confident Interval](https://www.mathsisfun.com/data/confidence-interval.html)

```python
tips.groupby('sex')['tip'].describe()
```

```python
sns.barplot?
```

```python
sns.barplot(data=tips, x='sex', y='tip')
```

```python
sns.barplot(data=tips, x='day', y='tip')
```

```python
sns.barplot(data=tips, x='day', y='tip', estimator=np.sum)
```

```python
sns.countplot(data=tips, x='day')
```

## sns.boxplot

```python
tips.groupby('day')['total_bill'].describe()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
```

## sns.stripplot

```python
sns.stripplot(data=tips, x='day', y='tip')
```

```python
sns.stripplot(data=tips, x='day', y='tip', hue='sex')
```

## sns.catplot

```python
sns.catplot(kind='bar', data=tips, col='day', x='sex', y='tip', estimator=np.mean, errwidth=0)
```

# Usage with matplotlib

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', ax=axes[0])
sns.countplot(data=tips, x='sex', ax=axes[1])
```
