# Data Visualization with Pandas

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Series.plot
- `line` : line plot (default)
- `bar` : vertical bar plot
- `barh` : horizontal bar plot
- `hist` : histogram
- `box` : boxplot
- `kde` : Kernel Density Estimation plot
- `density` : same as 'kde'
- `area` : area plot
- `pie` : pie plot

```python
pd.Series.plot?
```

```python
tips['total_bill'].plot(kind='hist', bins=15)
```

```python
tips['tip'].plot(kind='kde')
```

```python
tips['tip'].plot(kind='box')
```

# DataFrame.plot
  - bar
  - hist
  - box
  - kde
  - area
  - pie

```python
tips.plot(kind='scatter', x='total_bill', y='tip')
```

```python
tips.plot(kind='hexbin', x='total_bill', y='tip', gridsize=20)
```

```python
mean_tips = tips.groupby('day').mean()
mean_tips
```

```python
mean_tips.plot(kind='bar')
```

```python
mean_tips[['total_bill', 'tip']].plot(kind='bar', stacked=True)
```

```python
mean_tips['total_bill'].plot(kind='pie')
```

```python
mean_tips.plot(kind='pie', subplots=True, figsize=(12,4))
```

```python
tips.plot(kind='box')
```

# Plot Formatting
  - figsize
  - title
  - legend
  - xlim
  - ylim
  - xlabel
  - ylabel

```python
mean_tips[['total_bill', 'tip']].plot(kind='bar', figsize=(12,8), title="title", ylabel='$', ylim=(0, 25))
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(8,4))

axes[0].set_title('Total Bill')
tips['total_bill'].plot(kind='hist', ax=axes[0])

axes[1].set_title('Tip')
tips['tip'].plot(kind='hist', ax=axes[1])
```
