```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4]
y = [4, 1, 3, 2]
plt.plot(x, y)


# สำหรับคนที่ใช้ jupyter notebook แล้วกราฟไม่แสดงออกมาให้ใช้คำสั่ง plt.show()
# plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 100)
y1 = x ** 2
y2 = x ** 3

print(x[:5], x[-5:])
print(y1[:5], y1[-5:])
print(y2[:5], y2[-5:])
```

```python
plt.title('Chart Title')

plt.xlabel('X axis label')
plt.ylabel('Y axis label')

plt.grid()

plt.xlim((0, 5))
plt.ylim((0, 150))

plt.plot(x, y1, label='x^2')
plt.plot(x, y2, label='$x^3$')
plt.legend()
```

# Figure

![](https://matplotlib.org/stable/_images/anatomy.png)

## Figure (รูปทั้งหมด)
The whole figure. The figure is a container of all the child `Axes`, and `Artists` (titles, figure legends, etc), and the canvas. A figure can contain any number of Axes, but will typically have at least one.

## Axes (ระนาบ)
This is what you think of as 'a plot', it is the region of the image with the data space. The Axes contains two (or three in the case of 3D) `Axis` objects **Be aware of the difference between `Axes` and `Axis`**

The Axes class and its member functions are the primary entry point to working with the OO interface.

## Axis (แกน)
These are the number-line-like objects. They take care of setting the graph limits and generating the ticks (the marks on the axis) and ticklabels (strings labeling the ticks).

# The object-oriented interface and the pyplot interface
There are two ways to use Matplotlib:

- Explicitly create figures and axes, and call methods on them (the "object-oriented (OO) style").
- Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting.

## Figure

```python
fig = plt.figure()
# fig.add_axes([left, bottom , width, height])
ax1 = fig.add_axes([0.0, 0.0, 1.0, 1.0])
ax2 = fig.add_axes([0, 1.1, 1.0, 1.0])
```

```python
fig = plt.figure()

# fig.add_axes([left, bottom , width, height])
ax1 = fig.add_axes([0   ,0 ,1 , 1])
ax2 = fig.add_axes([1.1 ,0 ,1 , 1])

ax1.plot(x, y1)
ax1.set_title('y1 = x^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.grid()

ax2.plot(x, y2)
ax2.set_title('$y_2 = x^3$')
ax2.set_xlabel('x')
ax2.set_ylabel('y1')
```

# Color, Line Width, Line Style and Marker
https://matplotlib.org/stable/tutorials/introductory/pyplot.html#controlling-line-properties

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
```

### Color & Line width

```python
fig, axs = plt.subplots(figsize=(12,3), nrows=1, ncols=4)

axs[0].plot(x, y, color="black")
axs[1].plot(x, y, color="red", linewidth=2)
axs[2].plot(x, y, color="#8B008B", lw=3)
axs[3].plot(x, y, color="green", lw=0.5)
```

### Line Style

```python
fig, axs = plt.subplots(figsize=(12,3), nrows=1, ncols=4)

# possible linestype options are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
axs[0].plot(x, y, color="green", lw=3, linestyle='-')
axs[1].plot(x, y, color="green", lw=3, ls='-.')
axs[2].plot(x, y, color="green", lw=3, ls=':')
axs[3].plot(x, y, color="green", lw=3, ls='--')
```

### Marker

```python
fig, axs = plt.subplots(figsize=(12,3), nrows=1, ncols=4)

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axs[0].plot(x, y, color="blue", lw=1, ls='-', marker='+')
axs[1].plot(x, y, color="blue", lw=0, ls='--', marker='o')
axs[2].plot(x, y, color="blue", lw=3, ls='-', marker='s')
axs[3].plot(x, y, color="blue", lw=3, ls='', marker='3', markersize=10)
```

```python
fig, axs = plt.subplots(figsize=(12,3), nrows=1, ncols=4)

# marker size and color
axs[0].plot(x, y, color="purple", lw=1, ls='-', marker='o', markersize=2)
axs[1].plot(x, y, color="purple", lw=1, ls='-', marker='o', markersize=4)
axs[2].plot(x, y, color="purple", lw=1, ls='-', marker='o', markersize=8, 
            markerfacecolor="red")
axs[3].plot(x, y, color="purple", lw=1, ls='-', marker='s', markersize=8, 
            markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

# Annotating Figure

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html#matplotlib.axes.Axes.annotate

```python
x = np.arange(10)
y = x ** 2
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.annotate('hello', xy=(2, 40))
ax.annotate('hi', xy=(6, 10))
```

```python
x = np.arange(10)
y = x ** 2

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.annotate(
    'mytext', 
    xy=(9, 80),                 # หัวลูกศร
    xytext=(0.8, 0.2),         # ตำแหน่งตัวอักษร
    textcoords='axes fraction',
    arrowprops={'facecolor': 'black', 'shrink': 0.05},
    fontsize=16,
)

# ทดลองเปลี่ยนหัวลูกศรไปด้านขวาบน
# ทดลองย้ายตัวอักษรไปมุมขวาล่าง
```
